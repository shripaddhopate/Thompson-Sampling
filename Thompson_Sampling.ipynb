```python
data = pd.read_csv("Ads_CTR_Optimisation.csv")
data
```

```python
data.head(10)
```

```python
import random
N = 800
d = 10
ad_selected= []
numberOfRewards1 = [0]*d
numberOfRewards0 = [0]*d
total_reward = 0

for n in range(0,N):
  ad=0
  max_random = 0
  for i in range(0,d):
    random_beta = random.betavariate(numberOfRewards1[i] +1,numberOfRewards0[i] +1)

    if random_beta>max_random:
      max_random = random_beta
      ad=i
    ad_selected.append(ad)
    reward = data.values[n,ad]
    if reward == 1:
      numberOfRewards1[ad]+=1
    else:
      numberOfRewards0[ad]+=1 
    total_reward+= reward


```

```python
plt.hist(ad_selected)
plt.title("HISTOGRAM OF SELECTED AD")
plt.xlabel('Ads')
plt.ylabel('No of times The ad selected')
```
